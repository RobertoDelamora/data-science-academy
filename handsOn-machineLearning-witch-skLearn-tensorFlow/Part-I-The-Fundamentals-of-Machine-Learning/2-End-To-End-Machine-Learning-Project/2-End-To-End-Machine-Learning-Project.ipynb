# Predicting Housing Prices

Neste capítulo, será definido e aplicado o primeiro algorítimo de Machine Learning responsável por realizar predições em dados reais: dataset _California Housing Prices_. Este conjunto de dados provém do census da California de 1990 e traz dados específicos de pequenos "distritos" (ou Block Groups), cada qual contendo atributos como _population_, _median income_, _median housing price_, entre outros. * Visão holística sobre o problema de negócio;

## Conhecendo o Problema

Dentro deste contexto, o objetivo do projeto é treinar um algorítimo de Machine Learning para prever a mediana do preço das casas de acordo com os dados do census da California. A variável target é a *median_housing_price*. A ideia é utilizar o resultado deste modelo para alimentar um **outro** algorítimo de Machine Learning que, por sua vez, irá determinar se vale ou não a pena investir na região.

Através da análise do problema de negócio e das saídas esperadas, é possível concluir que trata-se de um modelo de **Aprendizagem Supervisionada** (saídas estão disponíveis e serão fornecidas) e também de **Regressão** (predição de valores numéricos/contínuos), mais especificamente uma **Regressão Multivariada** (múltiplas _features_ utilizadas para predição). Além disso, o contexto diz que não há um fluxo contínuo de dados e, portanto, ajustes em tempo real não se fazem necessário o que, dessa forma, caracteriza um modelo de **Batch Learning**.

## Coletando os Dados 

Após a apresentação dos conceitos teóricos e do problema de negócio alvo desta primeira análise, os primeiros passos para a implementação do algorítimo de Machine Learning passam pela criação de uma **função** que automaticamente realiza o download de conjuntos de dados da internet.

Esta metodologia é extremamente útil em casos onde os dados mudam rapidamente. Dessa forma, um pequeno script pode ser responsável por facilmente coletar o último conjunto de dados existentes de determinada fonte. Inclusive isto pode ser feito periodicamente em intervalos pre-definidos. Uma outra utilidade deste tipo de abordagem diz respeito a facilidade que se tem caso seja necessário baixar conjuntos de dados em diversas máquinas.

_Em projetos reais, o ideal é transformar esta função em um arquivo com extensão _.py_, porém, para fins didáticos, ela pode ser tranquilamente implementada neste jupyter notebook._

````python
# Importando bibliotecas, definindo variáveis e função para extração do dataset
import os
import tarfile # em casos de download de arquivos .tgz (linux)
from six.moves import urllib

DOWNLOAD_ROOT = 'https://raw.githubusercontent.com/ageron/handson-ml/master/' # Define url root onde encontram-se os arquivos
HOUSING_PATH = os.path.join('datasets', 'housing') # Pasta onde será(ão) organizado(s) o(s) arquivo(s)
HOUSING_URL = DOWNLOAD_ROOT + 'datasets/housing/housing.csv' # Parte específica da url para baixar arquivo específico

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    """
    Dada uma URL, esta função é responsável por realizar o download, extração e organização de datasets automaticamente
    
    INPUTS: 
    housing_url = URL no qual se encontra o arquivo a ser baixado.
    housing_path = caminho dentro do SO onde o arquivo será armazenado.
    """
    if not os.path.isdir(housing_path): # Se não há pasta com o nome específicado, cria uma no SO.
        os.makedirs(housing_path) 
    if 'tgz' in housing_url: # Se trata-se de um arquivo .tgz (Linux)
        try: 
            filename = 'housing'
            extension = '.tgz'
            tgz_path = os.path.join(housing_path, filename+extension)
            urllib.request.urlretrieve(housing_url, tgz_path)
            housing_tgz = tarfile.open(tgz_path)
            housing_tgz.extractall(path=housing_path)
            housing_tgz.close
        except tarfile.ReadError:
            print('Erro de Leitura: Não foi possível abrir o arquivo.')
        except urllib.error.URLError:
            print('Erro HTTP: URL não encontrada (Erro 404).')
    elif 'csv' in housing_url: # Se trata-se de um arquivo .csv
        try:
            filename = 'housing' # Nome do arquivo
            extension = '.csv' # Extensão
            csv_path = os.path.join(housing_path, filename+extension) # Mapeando caminho no SO
            urllib.request.urlretrieve(housing_url, csv_path) # Acessa link para download
        except urllib.error.URLError:
            print('Erro HTTP: URL não encontrada (Erro 404).')
        else:
            print(f'Download realizado com sucesso! Arquivo pode ser encontrado em: \n{housing_path+"/"+filename+extension}')
````

## Separando Conjuntos de Dados

Esta é uma etapa crucial que precede o recebimento do algorítimo de Machine Learning. Nela, são avaliadas todas as nuances e possíveis imperfeições nos dados que podem, de certa forma, atuar de modo a reduzir a eficiência do modelo. O trabalho nesta etapa interfere diretamente nos resultados.

### Conjuntos de Teste e de Treino

Com o auxílio do NumPy, será coletado aleatoriamente 20% do dataset

````python
import numpy as np

def split_train_test(data, test_ratio):
    """
    Realiza a separação de um conjunto de dados de acordo com o ratio pré-definido.
    
    INPUT:
    data: conjunto de dados [pandas.DataFrame]
    test_ratio: porcentagem de separação [float]
    
    OUTPUT:
    train_indices [pandas.DataFrame]
    test_indices [pandas.DataFrame]
    """
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
````

````python
# Apenas para entender melhor o que é realizado pela função, vamos printar o passo a passo
print(f'1 - Resultado de np.random.permutation: {np.random.permutation(len(df_housing))}')
print(f'2 - Resultado de test_set_size considerando 20% de ratio: {int(len(df_housing) * .20)}')
print(f'3 - Resultado de test_indices: {np.random.permutation(len(df_housing))[:int(len(df_housing) * .20)]}')
print(f'4 - Resultado de train_indices: {np.random.permutation(len(df_housing))[int(len(df_housing) * .20):]}')
````

Os retornos da função nada mais são que dois objetos DataFrames indexados (via .iloc) de acordo com os índices contidos em *test_indices* e *train_indices*.

````python
# Chamando a função que separa os dados
train_set, test_set = split_train_test(df_housing, .20)
print(f'{len(train_set)} train + {len(test_set)} test')
````

Chamando a função ```split_train_test(data, test_ratio)``` foi possível obter, randomicamente, duas variáveis, ou melhor, dois DataFrames divididos: **train_set** e **test_set**.

Apesar da implementação via NumPy ter funcionado corretamente, uma maneira bem mais fácil e eficiente de realizar a separação entre os conjuntos de teste e treino pode ser feita através do próprio **sklearn** através da função ```train_test_split()```.

[Documentação-train_test_split](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html)

````python
# Importando biblioteca e realizando o split dos dados
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df_housing, test_size=.2, random_state=42)
````

É importante ressaltar que o parâmetro random_state=42 funciona como uma garantia de que o conjunto de dados será separado da mesma maneira repetidas vezes, mantendo assim a consistência dos dados. Isto tem a ver com a semente randômica presente no NumPy (np.random.seed(42))

### Garantindo a Representatividade

Suponha que, conversando com os solicitantes do modelo, seja definido que o atributo **median_income** é extremamente relevante para a predição.

````python
# Observando novamente a distribuição de median_income
df_housing['median_income'].plot(kind='hist', figsize=(9, 6), grid=True, bins=50)
plt.xlabel('median_income')
plt.show()
````

O histograma acima indica que a maior concentração do atributo *median_income* está entre \$20,000 e \$50,000, porém alguns valores vão além de \$60,000. É muito importante que se tenha quantidades suficientes para cada _stratum_ (camada/classe), ou então a estimativa da importância dessa classe pode ser enviesado (deturpado).

    - Isto significa que não é adequado possuir muitas camadas/grupos/strata e cada um destes deve ser largo o suficiente.
    
Em busca da garantia de que este atributo seja representativo, será criada uma nova categoria ["income_cat"] que nada mais é do que um agrupamento dos valores contidos em median_income, aplicadas algumas limitações e transformações para reduzir a quantidade de camadas e aumentar a representatividade destas

````python
# Criando nova categoria
df_housing['income_cat'] = np.ceil(df_housing['median_income'] / 1.5) # Limitando valores
df_housing['income_cat'].where(df_housing['income_cat'] < 5, 5.0, inplace=True)

# Verificando
df_housing['income_cat'].hist(figsize=(10, 5));
````

Referência ```pd.DataFrame.where()```: https://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.where.html

Dessa forma, é possível aplicar uma amostragem estratificada na categoria *income_cat* através de uma função específica do sickit-learn: ```StratifiedShuffleSplit```

[Documentação-StratifiedShuffleSplit](http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.StratifiedShuffleSplit.html)

````python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=.2, random_state=42)
for train_index, test_index in split.split(df_housing, df_housing['income_cat']):
    strat_train_set = df_housing.loc[train_index]
    strat_test_set = df_housing.loc[test_index]

# Verificando
strat_test_set['income_cat'].value_counts() / len(strat_test_set)
````

````python
# Criação de strat_train_set e strat_test_set que recebem um slicing de df_housing de acordo com os índices dos arrays acima
for train_index, test_index in split.split(df_housing, df_housing['income_cat']):
    strat_train_set = df_housing.loc[train_index]
    strat_test_set = df_housing.loc[test_index]
strat_train_set.head()
````

````python
# Curiosidade: tamanho de cada um deles
print(f'Tamanho de strat_train_set: {strat_train_set.shape[0]} linhas e {strat_train_set.shape[1]} colunas.')
print(f'Tamanho de strat_test_set: {strat_test_set.shape[0]} linhas e {strat_test_set.shape[1]} colunas.')
````

````python
# São os mesmos valores dos conjuntos de treino e test (train_set e test_set) criados anteriormente
print(f'Tamanho de train_set: {train_set.shape[0]} linhas e {train_set.shape[1]} colunas.')
print(f'Tamanho de test_set: {test_set.shape[0]} linhas e {test_set.shape[1]} colunas.')
````

A grande jogada é que *strat_train_set* e *strat_test_set* são sets criados através de amostras estratificadas, cuja intenção é garantir uma maior representatividade dos conjuntos splitados, aumentando a eficiência e o realismo do modelo. A verificação final da representatividade é feita através da comparação da porcentagem das camadas (strata) da coluna **income_cat** com o conjunto original. 

````python
# Verificando porcentagens - dataset original
df_housing['income_cat'].value_counts() / len(df_housing)
````

````python
# Verificando porcentagens - strat_test_set
strat_test_set['income_cat'].value_counts() / len(strat_test_set)
````

A figura abaixo irá mostrar uma visão geral sobre a representatividade de cada um dos métodos de separação. Quanto mais próximos os valores do dataset original, mais confiável é a separação.

Dessa forma, é possível concluir que os conjuntos de testes gerados através das técnicas de estratificação são muito mais confiáveis no sentido de representar o conjunto como um todo. Por outro lado, as separações obtidas simplesmente por métodos randômicos podem ser levemente enviesadas, tornando um conjunto de testes, por exemplo, não representativo com relação ao modelo e prejudicando os resultados.

````python
# Após isso, é possível remover a coluna adicional criada
for set_ in (strat_train_set, strat_test_set):
    set_.drop('income_cat', axis=1, inplace=True)
````

### Transformações Próprias (CombAttrAdder)

O Sickit-Learn oferece uma série de transformações e é muito importante escrever as próprias funções para tarefas customizadas, como por exemplo, operações de limpeza e de combinação de atributos. O desejado é que essas transformações trabalhem com as demais funcionalidades oferecidas pelo Sickit Learn (como *pipelines*) e, uma vez que o Sickit funciona com *duck typing* (e não herança), basta criar a classe e implementar três métodos: ```fit()``` (returing self), ```transform()``` e ```fit_transform()``` (este último é obtido simplesmente pela adição da classe ```TransformerMixin```).

Adicionalmente, adicionando a classe ```BaseEstimator``` ganha-se os métodos extra ```get_params()``` e ```set_params()``` (úteis para tunar hyperparâmetros), desde que o BaseEstimator esteja sem os parâmetros ****args*** e **__kwargs__

````python
from sklearn.base import BaseEstimator, TransformerMixin

rooms_ix, bedrooms_ix, population_ix, household_ix = 3, 4, 5, 6 # Índices dos atributos a serem combinados

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room=True): # no *args or **kwargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self # nothing else 
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
        population_per_household = X[:, population_ix] / X[:, household_ix]
        if self. add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                        bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
````

````python
# Instanciando classe
attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
````

A classe criada nada mais é do que uma maneira (extremamente) elegante de realizar as mesmas operações feitas em **1.5.3 - Combinando Atributos**, com o adicional da possibilidade de contemplar um *hiperparâmetro* (argumento add_bedrooms_per_room) que, caso passado como **True**, realiza uma terceira combinação de atributos.

Este é o papel do hiperparâmetro em um modelo de Machine Learning: realizar operações, alterações e transformações no conjunto de dados de modo a verificar se o resultado do algorítimo se torna mais eficiente. Vamos verificar os resultados obtidos através da criação do objeto ```attr_adder``` e a consequente alteração em ```housing_extra_attribs```

### Feature Scaling (StandardScaler)

No geral, algorítimos de Machine Learning não são tão efetivos quando se tem atributos em diferentes escalas numéricas. Por exemplo, o atributo *total_rooms* vai de 6 até 39.320, enquanto o atributo *median_income* varia de 0 a 15.

Há duas maneiras de modificar os atributos de modo a trazê-los para a mesma escala:
* **min-max scaling**
* **standarization**

#### Dimensionamento: Min-Max

Também chamado de *normalization*.

Os valores são transformados e sua escala mudada para que, no final das contas, todos se encontrem em um range que varia de 0 a 1. Para tal, os dados de determinado atributo são subtraídos de seu valor mínimo e, em seguida, dividos pelo resultado do máximo menos o mínimo.

    - value = (value - min) / (max - min)
    
Para esta operação, o Scikit-learn oferece um transformer chamado ```MinMaxScaler```, tendo este um hiperâmetro chamado *feature_range* que permite alterar o range (se não desejar que seja entre 0 e 1). [Documentação-Oficial-Min-Max-Scaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html)

````python
from sklearn.preprocessing import MinMaxScaler

housing_scaler = housing.iloc[:, :-1].copy()
scaler = MinMaxScaler()
scaler.fit_transform(housing_scaler.loc[:, ['total_rooms', 'median_income']])
````

#### Dimensionamento: Standarization

O método de *Standarization* *(padronização)* atua da seguinte forma: primeiro é subtraído dos dados o valor da média (assim as Series possuem média igual a 0) e, em seguida dividi-se o valor pela variância. Dessa forma, os dados possuem variância unitário. Diferente do dimensionamento por **Min-Max**, *Standarization* não limita os valores para um range específico, o que pode ser um problema para alguns algorítmos.

    - Redes Neurais esperam um valor de input com range entre 0 e 1;

Standarization é muito menos afetado por outliers. por exemplo, imagine que o atributo *median_income* possua, erroneamente, um valor igual a 100. O dimensionamento Min-Max iria forçar o valor de 0 a 15 para 0 a 0.15, enquanto Standarization não iria proporcionar maiores alterações. O Scikit-Learn possui uma classe específica para tal: ```StandardScaler```. Link: [Documentação-Oficial-StandardScaler](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html)

````python
from sklearn.preprocessing import StandardScaler

housing_std_scaler = housing.loc[:, ['total_rooms', 'median_income']]
std_scaler = StandardScaler()
std_scaler.fit_transform(housing_std_scaler)
````

### Transformation Pipelines (ColumnTransformer)

Como pôde ser visto, há muitas transformações a serem executadas nos dados e em uma ordem correta. Felizmente, o Sickit-Learn fornece uma classe chamada ```Pipeline``` para ajudar nessa sequência de transformações. Abaixo será colocado um exemplo de uma criação de um Pipeline contendo algumas das transformações já realizadas nos *atributos numéricos*.

````python
# Importando bibliotecas
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

# Definindo pipeline
num_pipeline = Pipeline([
    ('imputer', Imputer(strategy='median')),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler()),
])

# Aplicando pipeline
housing_num = housing.drop('ocean_proximity', axis=1)
housing_num_tr = num_pipeline.fit_transform(housing_num)
````

O construtor do Pipeline rececbe uma lista de pares de *name/estimator* definindo a sequência dos passos. Todos os elementos estimator precisam ser transformers (precisam ter um método ```fit_transform()```), com excessão do último.

Seria interessante adicionar ao Pipeline um processo que separa atributos numéricos de categóricos. Scikit-Learn não fornece nada automático para realizar este processo, entretanto é possível criar uma classe específica para realizar este procedimento.

````python
from sklearn.base import BaseEstimator, TransformerMixin

class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values # Retorna o DataFrame indexado apenas com os atributos passados como argumento
````

````python
# Testando
selector = DataFrameSelector(housing.columns[:-1])
housing_numerical = selector.fit_transform(housing)
housing_numerical[0]
````

Dessa forma, é possível contruir um Pipeline eficiente para atributos numéricos e outro Pipeline para atributos categóricos.

````python
num_attribs = list(housing_num)
cat_attribs = ['ocean_proximity']

# Criando Pipeline para dados numéricos
num_pipeline = Pipeline([
    ('selector', DataFrameSelector(num_attribs)),
    ('imputer', Imputer(strategy='median')),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler()),
])

# Criando Pipeline para dados categóricos
cat_pipeline = Pipeline([
    ('selector', DataFrameSelector(cat_attribs)),
    ('cat_encoder', OneHotEncoder(sparse=False)),
])
````

Agora, o grande desafio é juntar os dois pipelines. O Scikit-Learn oferece uma classe chamada ```FeatureUnion``` para realizar este processo. Basta passar uma lista de *transformers* (ou os pipelines por completo).

````python
# Unindo Pipelines
from sklearn.pipeline import FeatureUnion

full_pipeline = FeatureUnion(transformer_list=[
                                             ('num_pipeline', num_pipeline),
                                             ('cat_pipeline', cat_pipeline),
                                             ])
````

### Cross-Validation

Uma maneira de avaliar os resultados do treinamento com o algorítimo ```DecisionTreeRegressor``` seria usar o método ```train_test_split``` para separar o conjunto de treinamento em um conjunto menor de treinamento e um conjuto de validação e, assim, treinar novamente o modelo e avaliar de acordo com o conjunto de validação gerado. Um pouco mais de trabalho mas nada impossível.

Uma alternativa é utilizar **cross-validation**, cuja ideia é criar *K-folds cross-validations*, ou seja, splits randômicos de novos subconjuntos (chamados *folds*) e realizar procedimentos sucessivos de treinamento e avaliação em uma quantidade **x** de vezes, sempre alternando o conjunto de avaliação e os de treinamento.

````python
# Utilizando cross validation com 10 subconjuntos
from sklearn.model_selection import cross_val_score

scores = cross_val_score(tree_reg, housing_prepared, housing_labels,
                        scoring='neg_mean_squared_error', cv=10)
tree_rmse_scores = np.sqrt(-scores)
````

````python
# Importando módulos necessários - Tratamento dos Dados
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import tarfile
from six.moves import urllib

# Importando módulos necessários - Preparação de Dados
from sklearn.model_selection import train_test_split
from sklearn.model_selection import StratifiedShuffleSplit
from sklearn.preprocessing import Imputer
from future_encoders import OneHotEncoder
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import FeatureUnion
from future_encoders import ColumnTransformer
from sklearn.pipeline import Pipeline

# Importando módulos necessários - Treinamento
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor

# Importando módulos necessários - Métricas e Validações
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error

# Importando módulos necessários - Sintonizando e Salvando o Modelo
from sklearn.model_selection import GridSearchCV
from scipy.stats import geom, expon
from sklearn.externals import joblib
````

Após a importação dos módulos, a primeira coisa a se fazer é ler o arquivo e **separar** o conjunto em datasets de treino e de teste. 

````python
# Definindo variáveis e função para extração do dataset
DOWNLOAD_ROOT = 'https://raw.githubusercontent.com/ageron/handson-ml/master/' # Define url root onde encontram-se os arquivos
HOUSING_PATH = os.path.join('datasets', 'housing') # Pasta onde será(ão) organizado(s) o(s) arquivo(s)
HOUSING_URL = DOWNLOAD_ROOT + 'datasets/housing/housing.csv' # Parte específica da url para baixar arquivo específico

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    """
    Dada uma URL, esta função é responsável por realizar o download, extração e organização de datasets automaticamente
    
    INPUTS: 
    housing_url = URL no qual se encontra o arquivo a ser baixado.
    housing_path = caminho dentro do SO onde o arquivo será armazenado.
    """
    if not os.path.isdir(housing_path): # Se não há pasta com o nome específicado, cria uma no SO.
        os.makedirs(housing_path) 
    if 'tgz' in housing_url: # Se trata-se de um arquivo .tgz (Linux)
        try: 
            filename = 'housing'
            extension = '.tgz'
            tgz_path = os.path.join(housing_path, filename+extension)
            urllib.request.urlretrieve(housing_url, tgz_path)
            housing_tgz = tarfile.open(tgz_path)
            housing_tgz.extractall(path=housing_path)
            housing_tgz.close
        except tarfile.ReadError:
            print('Erro de Leitura: Não foi possível abrir o arquivo.')
        except urllib.error.URLError:
            print('Erro HTTP: URL não encontrada (Erro 404).')
    elif 'csv' in housing_url: # Se trata-se de um arquivo .csv
        try:
            filename = 'housing' # Nome do arquivo
            extension = '.csv' # Extensão
            csv_path = os.path.join(housing_path, filename+extension) # Mapeando caminho no SO
            urllib.request.urlretrieve(housing_url, csv_path) # Acessa link para download
        except urllib.error.URLError:
            print('Erro HTTP: URL não encontrada (Erro 404).')
        else:
            print(f'Download realizado com sucesso! Arquivo pode ser encontrado em: \n{housing_path+"/"+filename+extension}')

# Chamando função
fetch_housing_data()

# Função para leitura de arquivo
def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path, 'housing.csv')
    return pd.read_csv(csv_path)

# Lendo arquivo
housing = load_housing_data()

# Verificando cabeçalho
housing.head()
````

````python
# Criando nova categoria
housing['income_cat'] = np.ceil(housing['median_income'] / 1.5) # Limitando valores
housing['income_cat'].where(housing['income_cat'] < 5, 5.0, inplace=True)

# Separando dataset
split = StratifiedShuffleSplit(n_splits=1, test_size=.2, random_state=42)
for train_index, test_index in split.split(housing, housing['income_cat']):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]

# Comparando
print(f'Shape do dataset original: {housing.shape}')
print(f'Shape de strat_train_set: {strat_train_set.shape}')
print(f'Shape de strat_test_set: {strat_test_set.shape}')

# Excluindo colunas adicionais
for set_ in (strat_train_set, strat_test_set):
    set_.drop('income_cat', axis=1, inplace=True)
    
# Criando cópia
housing = strat_train_set.drop('median_house_value', axis=1)
housing_labels = strat_train_set['median_house_value'].copy()

print(f'Shape de housing modificado: {housing.shape}')
print(f'Shape de housing_labels: {housing_labels.shape}')
````

````python
# Classe responsável por combinar alguns atributos do dataset

rooms_ix, bedrooms_ix, population_ix, household_ix = 3, 4, 5, 6

class CombinadorAtributos(BaseEstimator, TransformerMixin):
    """
    Classe criada para combinar atributos e criar novas features de acordo com as informações contidas no Dataset.
    Atributos criados:
        - 
    """
    def __init__(self, add_bedrooms_per_room=True, add_rooms_per_population=True):
        """
        Método construtor: define hiperparâmetros, neste caso, combinações entre atributos diferentes
            - add_bedrooms_per_room
            - add_rooms_per_population
        """
        self.add_bedrooms_per_room = add_bedrooms_per_room
        self.add_rooms_per_population = add_rooms_per_population
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
        population_per_household = X[:, population_ix] / X[:, household_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            if self.add_rooms_per_population:
                rooms_per_population = X[:, rooms_ix] / X[:, population_ix]
                return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room,
                            rooms_per_population]
            else:
                return np.c_[X, rooms_per_household, population_per_household, bedrooms_per_room]
        if self.add_rooms_per_population:
            rooms_per_population = X[:, rooms_ix] / X[:, population_ix]
            return np.c_[X, rooms_per_household, population_per_household, rooms_per_population]
        else:
            return np.c_[X, rooms_per_household, population_per_household]
````
